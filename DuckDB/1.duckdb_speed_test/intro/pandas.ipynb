```python
from pathlib import Path
import pandas as pd
```

```python
input_file = Path("../data") / "companies_sorted.parquet"
```

```python
%%time
df = pd.read_parquet(input_file)
# Count number of rows in parquet file
print(f"{df.shape[0]} rows in parquet file")
df.head()
```

```python
print(df.columns)
df = df[df["year founded"].notnull() & df["country"].notnull()]
```

```python
%%time
# Read specific columns from parquet file and sort by year founded
df1 = df[["name", "year founded", "country"]] \
    .sort_values("year founded", ascending=True)
df1.head()
```

```python
%%time
# Group by industry and total number of estimated employees
df2 = (
    df.groupby("industry")["current employee estimate"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
df2.head()
```

```python
%%time
# Group by country and total number of estimated employees
df3 = (
    df.groupby("country")["current employee estimate"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
df3.head()
```

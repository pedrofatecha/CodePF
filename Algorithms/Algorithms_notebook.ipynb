```python

class recursiveformulas:
    def recursivesum(N):
            if N==1:
                return 1
            else:
                return N+recursiveformulas.recursivesum(N-1)

    def tail(n):
        if n==0:
            return n
        print(n)
        recursiveformulas.tail(n-1)


    def head(n):
        if n==0:
            return n
        recursiveformulas.head(n-1)
        print(n)


    def factorial(n):
        if n==1:
            return 1
        return n*recursiveformulas.factorial(n-1)

    def fibonacci(n):
        if n==0: return 0
        if n==1: return 1
        fib1= recursiveformulas.fibonacci(n-1)
        fib2= recursiveformulas.fibonacci(n-2)
        res=fib1+fib2

        return res

    def hanoi(n,rod_from,rod_middle,rod_to):

        if n==1:
            print("plate 1 from %s to %s"%(rod_from,rod_to))
            return

        recursiveformulas.hanoi(n-1,rod_from,rod_to,rod_middle)
        print("plate %s from %s to %s"%(n,rod_from,rod_to))

        recursiveformulas.hanoi(n-1,rod_middle,rod_from,rod_to)

    def binary(array, item, left, right):
        if right<left:
            return -1;
        middle= left+(right-left)//2
        print("middle index: %s"%(middle))

        if array[middle]==item:
            return middle

        elif array[middle]>item:
            print("Checking items on the left...")
            return recursiveformulas.binary(array, item, left, middle-1)
        
        elif array[middle]<item:
            print("Checking items on the right...")
            return recursiveformulas.binary(array, item, middle+1, right)

    
if __name__ == "__main__":    #re=recursivesum(2)
    print(recursiveformulas.binary([1,4,5,6,7,8,9,10,11], 11, 0, 7))
```

## Sorting Algorithm

```python

class insertionsort:
    def insert(A_list, j):
      key=A_list[j]
      i=j-1
      while i>=0 and A_list[i]>key:
          A_list[i+1]=A_list[i]
          i-=1
      A_list[i+1]=key

    def insert_des(A_list, j):
      key=A_list[j]
      i=j-1
      while i>=0 and A_list[i]<key:
          A_list[i+1]=A_list[i]
          i-=1
      A_list[i+1]=key

    def sort_insert(A_list):
        for n in range(0, len(A_list)):
            insertionsort.insert(A_list, n)
        return A_list

    def sort_insert_desc(A_list):
        for n in range(0, len(A_list)):
            insertionsort.insert_des(A_list, n)
        return A_list

if __name__ == "__main__":
    A_list = [5, 6, 1, 2, 4, 3]
    
    print("Before sorting: ", A_list)
    sorted_list = insertionsort.sort_insert_desc(A_list)
    print("After sorting: ", sorted_list)
    # Example usage of recursiveformulas

```
